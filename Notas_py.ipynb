# rjust()
Agrega los caracteres filtrados al izquiero de la varialbe. Recibe 2 par√°metros, la cantidad de car√°cteres a llenar (incluyendo el texto) y el caracter con el que se llenar√°n los espacios:

# ljust()
Agrega los caracteres filtrados al lado derecho del texto. Recibe 2 par√°metros, la cantidad de car√°cteres a llenar (incluyendo el texto) y el caracter con el que se llenar√°n los espacios:

````python
print(curso.ljust(50, '*'))
````

# replace()

# count()

> Se recomienda siempre pasar el texto a min√∫sculas con `lower()` antes de reemplazar

````python
print(curso.count('o'))
````

# find()
Encuentra y retorna la posici√≥n de la primera similutd del caracter buscado. Recibe 1 par√°metro, caracter a buscar.

````python
print(curso.find('p'))
print(curso.find('z'))
````

# lstrip()
Formatea el texto al lado derecho.

````python
espacios_derecha: str = '      Originalmente, tengo espacios a mi der'
print('Texto original: ', espacios_derecha)
print('Texto seteado: ', espacios_derecha.lstrip())
````

#  rstrip()
Formatea el texto al lado izq.

````python
espacios_izq: str = 'Originalmente tengo espacios a mi izq         '
print('Texto original: ', espacios_izq)
print('Texto seteado : ', espacios_izq.rstrip())
````

# strip()
Formatea el texto a ambos lados.

````python
espacio_ambos: str = '       Que desastre que se llene un formulario as√≠        '
print('Texto original: ', espacio_ambos)
print('Texto seteado : ', espacio_ambos.strip())
````
